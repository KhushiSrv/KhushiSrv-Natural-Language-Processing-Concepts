```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
```

```python
#Tokenization
def tokenize_text(text):
    sentences = sent_tokenize(text)
    tokenized_text = [word_tokenize(sentence) for sentence in sentences]
    return tokenized_text
text_to_tokenize = "The sky is blue and the grass is green."

tokenized_result = tokenize_text(text_to_tokenize)
for sentence_tokens in tokenized_result:
    print(sentence_tokens)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
#Lemmationation
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text

text_to_lemmatize = "The sky is blue and the grasses is green."
lemmatized_text = lemmatize_text(text_to_lemmatize)
print("Original text:", text_to_lemmatize)
print("Lemmatized text:", lemmatized_text)
```

```python
def ngram_probabilities(text, n):
    words = text.split()
    ngrams_list = list(ngrams(words, n))
    ngram_probabilities = Counter(ngrams_list)
    total_ngrams = len(ngrams_list)

    for ngram, count in ngram_probabilities.items():
        ngram_probabilities[ngram] = count / total_ngrams

    return ngram_probabilities
```

```python
import nltk
import pandas as pd

sentence = "King Krishnadevaraya loved horses and had the best collection of horse breeds in the Kingdom. Well, one day, a trader came to the King and told him that he had brought with him a horse of the best breed in Arabia.He invited the King to inspect the horse. King Krishnadevaraya loved the horse; so the trader said that the King could buy this one and that he had two more like this one, back in Arabia that he would go back to get. The King loved the horse so much that he had to have the other two as well. He paid the trader 5000 gold coins in advance. The trader promised that he would return within two days with the other horses.Two days turned into two weeks, and still, there was no sign of the trader and the two horses. One evening, to ease his mind, the King went on a stroll in his garden. There he spotted Tenali Raman writing down something on a piece of paper. Curious, the King asked Tenali what he was jotting down.Tenali Raman was hesitant, but after further questioning, he showed the King the paper. On the paper was a list of names, the King’s being at the top of the list. Tenali said these were the names of the biggest fools in the Vijayanagara Kingdom!As expected, the King was furious that his name was at the top and asked Tenali Raman for an explanation. Tenali referred to the horse story, saying the King was a fool to believe that the trader, a stranger, would return after receiving 5000 gold coins.Countering his argument, the King then asked, what happens if/when the trader does come back? In true Tenali humour, he replied saying, in that case, the trader would be a bigger fool, and his name would replace the King’s on the list!"
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(sentence)
```

```python
from nltk.tag import pos_tag
tagged = pos_tag(tokens)
tagged
```

```python
from nltk.chunk import RegexpParser

grammar = r"""
  NP: {<DT>?<JJ>*<NN>}
      {<NNP>+}
"""

chunk_parser = RegexpParser(grammar)
chunked = chunk_parser.parse(tagged)
```

```python
noun_phrases = []
for subtree in chunked.subtrees():
    if subtree.label() == 'NP':
        noun_phrase = ' '.join(word for word, tag in subtree.leaves())
        noun_phrases.append(noun_phrase)

print("Noun Phrases:", noun_phrases)
```

SEMANTIC RELATIONSHIP


```python
from itertools import combinations

import nltk
from nltk.corpus import wordnet

# Download WordNet data
nltk.download('wordnet')
```

```python
def get_semantic_relationship(word1, word2):
    # Get synsets (sets of synonyms) for each word
    synsets1 = wordnet.synsets(word1)
    synsets2 = wordnet.synsets(word2)

    # Initialize maximum similarity
    max_similarity = -1

    # Iterate through each pair of synsets and find their path similarity
    for synset1 in synsets1:
        for synset2 in synsets2:
            similarity = synset1.path_similarity(synset2)
            if similarity is not None and similarity > max_similarity:
                max_similarity = similarity

    return max_similarity
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def analyze_sentence(sentence):
    words = sentence.split()
    relationships = {}

    # Find semantic relationships between all pairs of words
    for word1, word2 in combinations(words, 2):
        relationship = get_semantic_relationship(word1, word2)
        relationships[(word1, word2)] = relationship

    return relationships

def plot_relationships(relationships):
    G = nx.Graph()

    # Add edges to the graph with weights representing relationships
    for pair, relationship in relationships.items():
        word1, word2 = pair
        G.add_edge(word1, word2, weight=relationship)

    # Draw the graph
    pos = nx.spring_layout(G)  # Positions for all nodes
    nx.draw(G, pos, with_labels=True, font_weight='bold')

    # Add labels to the edges
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    # Show the plot
    plt.show()

sentence = "The robot explored the abandoned city"
relationships = analyze_sentence(sentence)
print("Semantic relationships in the sentence:")
for pair, relationship in relationships.items():
    print(f"{pair}: {relationship}")

plot_relationships(relationships)
```
